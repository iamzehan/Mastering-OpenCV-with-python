````python
# drawing a triangle
image = np.zeros((400,400,3), dtype='uint8')
image[:] = colors['white']

pts = np.array([[250,5],[220,80],[280,80]],np.int32) # top, left, right
pts = pts.reshape((-1,1,2)) # (number_vertex, 1, 2)
print(pts.shape)
image = cv2.polylines(image,[pts],True,colors['blue'],2)
pts = np.array([[100,200],[50,275],[150,275]],np.int32) # top left right
pts = pts.reshape((-1,1,2)) # (number_vertex, 1, 2)
image = cv2.fillPoly(image,[pts],colors['red'])
plt.grid()
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Drawing a triangle")
plt.show()
````

````python
# drawing a square
image = np.zeros((400,400,3), dtype='uint8')
image[:] = colors['white']

# non-filled square
pts = np.array([[100,50],[100,150],[200,150],[200,50]],np.int32) # top, left, right
pts = pts.reshape((-1,1,2)) # (number_vertex, 1, 2)
print(pts.shape)
image = cv2.polylines(image,[pts],True,colors['blue'],2)

#circle inside the square 
image = cv2.circle(image, (150,100),50, colors['red'],2)

# filled square
pts = np.array([[200,200],[200,300],[300,300],[300,200]],np.int32) # top left right
pts = pts.reshape((-1,1,2)) # (number_vertex, 1, 2)
image = cv2.fillPoly(image,[pts],colors['red'])

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Drawing a Square")
plt.show()
````

### **Now let's look into shift parameter**
___
shift parameter allows you to specify the number of the fractional bits (on the right of the decimal point) In the end the real point coordinates are calculated as follows:

$Point(x,y)  Point2f(x^{-shift},y^{-shift})$

````python
image = np.zeros((400,400,3), dtype='uint8')
# image[:] = colors['white']
shift = 2
factor = 2**shift
image = cv2.circle(image, (int(round(299.99 * factor)),int(round(299.99*factor))), 100*factor, colors['red'], 2, shift=shift)
image = cv2.circle(image,(299,299),100, colors['green'],2)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Drawing circle with shift parameter")
plt.show()
````

#### **Writing Text**
___

```python:
cv2.putText(image, text, pt, font_style,font_size, color, thickness, line_type)
```

Available fonts are:

* FONT_HERSHEY_SIMPLEX = 0
* FONT_HERSHEY_PLAIN = 1
* FONT_HERSHEY_DUPLEX = 2
* FONT_HERSHEY_TRIPLEX = 3
* FONT_HERSHEY_COMPLEX = 4
* FONT_HERSHEY_COMPLEX_SMALL = 5
* FONT_HERSHEY_SCRIPT_SIMPLEX = 6
* FONT_HERSHEY_SCRIPT_COMPLEX = 7

````python
image = np.zeros((600,1000,3), dtype='uint8')
image[:] = colors['white']

fonts=[
  'FONT_HERSHEY_SIMPLEX',
  'FONT_HERSHEY_PLAIN',
  'FONT_HERSHEY_DUPLEX',
  'FONT_HERSHEY_TRIPLEX',
  'FONT_HERSHEY_COMPLEX',
  'FONT_HERSHEY_COMPLEX_SMALL',
  'FONT_HERSHEY_SCRIPT_SIMPLEX',
  'FONT_HERSHEY_SCRIPT_COMPLEX'
]
index_colors=[value for key,value in colors.items()]
position=(50,50)
fontsize=0.5

for i in range(0,8):
  image = cv2.putText(image, fonts[i], position, i, fontsize, index_colors[i], 1, cv2.LINE_AA)
  position = (position[0],position[1]+50)
  fontsize+=0.1
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Drawing Texts")
plt.show()
````

````python
image = np.zeros((600,1000,3), dtype='uint8')
# image[:] = colors['white']

font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1.5
thickness = 2
text = 'abcdefghijklmnopqrstuvwxyz'
circle_radius = 10
ret, baseline = cv2.getTextSize(text, font, font_scale, thickness)

text_width, text_height = ret

# center the text in the image
text_x = int(round((image.shape[1] - text_width)/2))
text_y = int(round((image.shape[0]- text_height)/2))

# Draw this point for reference:
# image=cv2.circle (image, (text_x, text_y), circle_radius, colors['green'], -1)

# Draw bounding box rectangle around the text
image = cv2.rectangle(image, (text_x, text_y+baseline), (text_x+text_width-thickness, text_y-text_height), colors['green'], thickness)

# Draw the circles defining the rectangle

image = cv2.circle(image, (text_x, text_y+baseline), circle_radius, colors['red'], -1)
image = cv2.circle(image, (text_x+text_width - thickness, text_y-text_height),circle_radius, colors['cyan'],-1)

image = cv2.putText(image, text, (text_x, text_y), font, font_scale, colors['white'], thickness) 

# plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
# plt.title("cv2.getTextSize() + cv2.putText()")
# plt.show()
cv2_imshow(image)
````

#### **Advanced Drawings**

````python
import datetime
import math
image = np.zeros((640,640,3), dtype='uint8')
image[:] = colors['white']

hours_orig = np.array(
    [(620, 320), (580, 470), (470, 580), (320, 620), (170, 580), (60, 470), (20, 320), (60, 170), (169, 61), (319, 20),
     (469, 60), (579, 169)])

hours_dest = np.array(
    [(600, 320), (563, 460), (460, 562), (320, 600), (180, 563), (78, 460), (40, 320), (77, 180), (179, 78), (319, 40),
     (459, 77), (562, 179)])

for i in range(0, 12):
    image=cv2.line(image, tuple(hours_orig[i]), tuple(hours_dest[i]), colors['black'], 3)

image=cv2.circle(image, (320, 320), 310, colors['dark_gray'], 8)
image=cv2.rectangle(image, (150, 175), (490, 270), colors['dark_gray'], -1)
image=cv2.putText(image, "Mastering OpenCV 4", (150, 200), 1, 2, colors['light_gray'], 1, cv2.LINE_AA)
image=cv2.putText(image, "with Python", (210, 250), 1, 2, colors['light_gray'], 1, cv2.LINE_AA)

image_original = image.copy()
# Get current date:
date_time_now = datetime.datetime.now()

# Get current time from the date:
time_now = date_time_now.time()

# Get current hour-minute-second from the time:
hour = math.fmod(time_now.hour, 12)
minute = time_now.minute
second = time_now.second

# Get the hour, minute and second angles:
second_angle = math.fmod(second * 6 + 270, 360)
minute_angle = math.fmod(minute * 6 + 270, 360)
hour_angle = math.fmod((hour*30) + (minute/2) + 270, 360)

# Draw the lines corresponding to the hour, minute and second needles:
second_x = round(320 + 310 * math.cos(second_angle * 3.14 / 180))
second_y = round(320 + 310 * math.sin(second_angle * 3.14 / 180))
image=cv2.line(image, (320, 320), (second_x, second_y), colors['blue'], 2)

minute_x = round(320 + 260 * math.cos(minute_angle * 3.14 / 180))
minute_y = round(320 + 260 * math.sin(minute_angle * 3.14 / 180))
image=cv2.line(image, (320, 320), (minute_x, minute_y), colors['blue'], 8)

hour_x = round(320 + 220 * math.cos(hour_angle * 3.14 / 180))
hour_y = round(320 + 220 * math.sin(hour_angle * 3.14 / 180))
image=cv2.line(image, (320, 320), (hour_x, hour_y), colors['blue'], 10)

image=cv2.circle(image, (320, 320), 10, colors['dark_gray'], -1)


plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Advanced drawing")
plt.show()
````
